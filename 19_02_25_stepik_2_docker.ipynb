# "Как?" vs "Зачем?"
---

- Сначала "Зачем?" и только потом "Как?"
- Не стоит изучать технологию ради технологии - решайте типовые задачи, для которых данная технология была разработана
- Проблемы обучения по документации
- Важно смотреть на технологии в исторической ретроспективе
- Понимание исторического контекста и даст ответ на вопрос "Зачем?"

# Процесс как универсальная единица выполнения программного кода
---

- Запуск любой программы - это запуск отдельного процесса
  - Системные процессы (управляют работой ОС)
    - Systemd
    - SSHD
    - cron
  - Пользовательские процессы
    - Браузер
    - Редактор кода
    - Телеграм-клиент
    - Терминал
    - Запуск простого Python-скрипта интерпретатором
  - Фоновые процессы
    - Redis
    - NGINX
    - Postgres
    - pulseaudio
    - bluetoothd
- Процессы могут общаться между собой разными способами
- Процессу для работы нужны ресурсы ОС

# Переход от общения между процессами на одной машине к распределённым системам
---

- Некоторые процессы могут общаться между собой по сети
- При этом не суть важно, где именно эти процессы запущены
- У них могут быть IP-адреса и порты для доступа извне
- Когда вы думаете о процессах - можете предсавлять их всегда запущенными на разных хостах
- Такой подход позволит вам начать мыслить в рамках событийно-ориентированной микросервисной архитектуры

# Велосипеды и спорткары
---

- Современная разработка становится всё больше похожа на конструктор Лего
- Зачем писать что-то своё непрофессионально и на коленке, если есть готовые профессиональные инструменты
- Когда вы берёте профессиональный инструмент - вы автоматически получаете себе в проект профессиоанльных контрибьюторов
- Уметь конфигурировать разные сервисы выходит на первый план, по сравнению с написанием кода
- Нужно ориентироваться в зоопарке инструментов

# При чём тут Docker?
---

- Это способ управлять процессами изолированно, то есть так, будто они на разных машинах
- При этом, по сути, единственное требование - иметь docker engine на хосте
- Docker - это универсальная среда запуска процессов
- Идеальный способ тестировать (изучать) тот самый зоопарк
  - Разные сервисы
  - Разные версии одних и тех же сервисов

# А можно без докера?
---

- Можно. Раньше же как-то жили без автомобилей и самолётов.
- Виртуальные машины
- Выделенные сервера
- Установка прямо в систему
- Виртуальные окружения
- Podman
- Даже если решаешь не пользоваться докером - это должен быть осознанный, а не безальтернативный выбор

# Почему специалист, владеющий докером, стоит дороже?
---

- На примере Python-разработки (hh):
  - **10 879** вакансий «python»
  - **2 859** вакансий «python docker»
  - **7 099** вакансий «docker»

# Docker для работы
---

- Тестирование инструментов без "засорения" системы
  - Cool story про Postgres
- Упрощённая установка и настройка окружения (инфраструктуры)
  - Пример быстрого развёртывания Postgres, NATS, Redis одной командой
- Разработка без "у меня работает"
- Развёртывание приложений одной командой на удалённом сервере или даже серверах
- Тестирование приложения перед обновлением (версии Python, например)
- Жёсткая фиксация зависимостей для воспроизводимости
  - Частая проблема в ML: новые версии библиотек несовместимы со старыми
- Тестирование API без установки Postman, а также без регистрации и SMS
- Безопасное тестирование подозрительного кода

# Docker для учёбы и исследований
---

- Тестирование кода на разных версиях языка
- Запуск интерпретаторов разных языков для работы в интерактивном режиме
  - На примере Python: `docker run --rm -it python bash`
- Быстрое развёртывание чужого приложения для изучения архитектурных подходов
  - На примере телеграм-бота из курса `git clone https://github.com/MasterGroosha/advanced-telegram-bots.git`
- Эмуляция удалённого сервера с доступом по SSH для тренировки и без риска что-то сломать
  - Пример
- Изучение новых инструментов
  - Пример с Postgres `docker exec -it <postgres_name_container> psql -U <user_name>`
  - Redis
  - NATS
  - RabbitMQ
- Эксперименты с несколькими серверами
  - Ручками настроить доступы по цепочке серверов по SSH
- Тестирование ML-моделей
- Эксперименты с распределённой архитектурой
- Создание кластеров на одной машине
