# Как получить в свое распоряжение Linux, если у вас Windows?
---

- Установить основной ОС
- Установить дополнительной ОС
- Поставить WSL (Windows Subsystem for Linux)
- Приобрести VPS (удаленный сервер) под управлением Linux
- Установить виртуальную машину и поставить туда Linux
- Поднять докер-контейнер c линуксом
